```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import json
import pandas as pd

# 폴더 내의 모든 JSON 파일 경로 가져오기 ##########(각자 경로 설정하기)##########
folder_path = '/Volumes/Crucial X6/반려동물 피부 질환 데이터/01.데이터/1.Training/2.라벨링데이터/반려견/피부/일반카메라/유증상/'
os_listdir = os.listdir(folder_path)

# 빈 데이터 프레임 생성
df_list = []

# 각 폴더에 접근하여 내부의 JSON 파일을 가져오기
for sub_folder in os_listdir:
    sub_folder_path = os.path.join(folder_path, sub_folder)
    if os.path.isdir(sub_folder_path):  # 디렉토리인 경우에만 처리
        json_files = [f for f in os.listdir(sub_folder_path) if f.endswith('.json')]
        for json_file in json_files:
            with open(os.path.join(sub_folder_path, json_file), 'r') as f:
                data = json.load(f)
                # metaData
                
                breed = data['metaData']['breed']
                age = data['metaData']['age']
                gender = data['metaData']['gender']
                region = data['metaData']['region']
                lesions = data['metaData']['lesions']
                path = data['metaData']['Path']
                src_path = data['metaData']['src_path']
                label_path = data['metaData']['label_path']
                
                
                # 데이터 프레임에 추가할 딕셔너리 생성
                data_dict = {'breed': breed, 'age': age, 
                            'gender': gender, 'region': region, 'lesions': lesions, 
                            'path': path, 'src_path': src_path, 'label_path': label_path}
                
                # 데이터 프레임 생성 및 리스트에 추가
                df_list.append(pd.DataFrame([data_dict]))

        # 데이터 프레임 합치기
        df = pd.concat(df_list, ignore_index=True)

# # ImageDataGenerator를 사용하여 데이터 로드
# datagen = ImageDataGenerator(rescale=1./255)  # 이미지 스케일링을 위해 rescale 옵션 사용
# image_size = (224, 224)  # 이미지 크기 설정

# # flow_from_dataframe을 사용하여 이미지 데이터를 로드
# data_generator = datagen.flow_from_dataframe(
#     dataframe=df,
#     x_col='src_path',  # 이미지 경로 컬럼
#     y_col='label_path',  # 라벨(레이블) 경로 컬럼
#     target_size=image_size,  # 이미지 크기 설정
#     batch_size=32,  # 배치 크기 설정
#     class_mode='categorical',  # 분류 모드 설정
#     shuffle=True  # 데이터를 섞을지 여부 설정
# )
df
```
